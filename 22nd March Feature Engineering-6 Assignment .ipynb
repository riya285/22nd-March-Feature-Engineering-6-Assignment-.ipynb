```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.




To calculate the Pearson correlation coefficient between two variables, you need to follow these steps:

Calculate the mean (average) of both the time spent studying and the exam scores.
Calculate the difference between each data point and its respective mean for both variables.
Multiply the differences obtained in step 2 for each data point and sum up these products.
Calculate the square root of the sum of squared differences from step 3 for both variables.
Divide the sum of products from step 3 by the product of the square roots obtained in step 4.
Mathematically, the formula for the Pearson correlation coefficient (r) is:

�
=
∑
�
=
1
�
(
�
�
−
�
ˉ
)
(
�
�
−
�
ˉ
)
∑
�
=
1
�
(
�
�
−
�
ˉ
)
2
∑
�
=
1
�
(
�
�
−
�
ˉ
)
2
r= 
∑ 
i=1
n
​
 (x 
i
​
 − 
x
ˉ
 ) 
2
 ∑ 
i=1
n
​
 (y 
i
​
 − 
y
ˉ
​
 ) 
2
 
​
 
∑ 
i=1
n
​
 (x 
i
​
 − 
x
ˉ
 )(y 
i
​
 − 
y
ˉ
​
 )
​
 

Where:

�
n is the number of data points.
�
�
x 
i
​
  and 
�
�
y 
i
​
  are the individual data points for time spent studying and exam scores, respectively.
�
ˉ
x
ˉ
  and 
�
ˉ
y
ˉ
​
  are the means of the time spent studying and exam scores, respectively.
The Pearson correlation coefficient 
�
r ranges from -1 to 1. A positive value indicates a positive linear relationship, meaning that as one variable increases, the other tends to increase as well. A negative value indicates a negative linear relationship, meaning that as one variable increases, the other tends to decrease. A value closer to 0 indicates a weaker or no linear relationship between the variables.

Interpretation of the result:

If 
�
r is close to 1, there's a strong positive linear relationship between time spent studying and exam scores, implying that students who study more tend to achieve higher exam scores.
If 
�
r is close to -1, there's a strong negative linear relationship between time spent studying and exam scores, suggesting that students who study less tend to achieve higher exam scores.
If 
�
r is close to 0, there's little to no linear relationship between the two variables, meaning that the amount of time spent studying does not have a significant impact on exam scores.











Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.



To calculate Spearman's rank correlation coefficient for the given data on sleep duration and job satisfaction level, follow these steps:

Step 1: Rank the data for both variables separately.
Step 2: Calculate the differences between the ranks of the two variables.
Step 3: Square the rank differences.
Step 4: Sum up the squared rank differences.
Step 5: Use the formula to calculate the Spearman's rank correlation coefficient.

Assuming you have data for a sample of individuals:

Individual	Sleep Duration	Job Satisfaction
1	6	7
2	8	6
3	7	5
4	5	8
5	6	4
Step 1: Rank the data for both variables separately:

Individual	Sleep Duration (X)	Rank_X	Job Satisfaction (Y)	Rank_Y
1	6	3	7	3
2	8	5	6	2
3	7	4	5	1
4	5	2	8	5
5	6	3	4	1
Step 2: Calculate the differences between the ranks of the two variables:

Individual	Rank_X	Rank_Y	Rank Difference (d)	d^2
1	3	3	0	0
2	5	2	3	9
3	4	1	3	9
4	2	5	-3	9
5	3	1	2	4
Step 3: Square the rank differences.

Step 4: Sum up the squared rank differences: Sum of d^2 = 0 + 9 + 9 + 9 + 4 = 31.

Step 5: Use the formula to calculate the Spearman's rank correlation coefficient:

Spearman's Rank Correlation Coefficient (ρ) = 1 - (6 * Sum of d^2) / (N * (N^2 - 1))
Where N is the number of observations.

In this case, N = 5.
ρ = 1 - (6 * 31) / (5 * (5^2 - 1))
ρ = 1 - (186) / (120)
ρ = 1 - 1.55
ρ ≈ -0.55

Interpretation:
The calculated Spearman's rank correlation coefficient (ρ) is approximately -0.55. This value indicates a moderate negative monotonic relationship between the amount of sleep individuals get each night and their overall job satisfaction level. In other words, as the amount of sleep decreases, job satisfaction tends to decrease, and as the amount of sleep increases, job satisfaction tends to increase. The negative sign of the correlation coefficient suggests an inverse relationship, while the absolute value (0.55) indicates the strength of the correlation.







Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.


To calculate both the Pearson correlation coefficient and Spearman's rank correlation, you will need to have the data for both variables: the number of hours of exercise per week and the corresponding Body Mass Index (BMI) for each of the 50 participants. Once you have the data, you can proceed with the calculations.

Here are the steps to calculate both correlation coefficients and a comparison of the results:

Step 1: Organize the Data

Let's assume you have the data in two lists: hours_of_exercise and bmi. Each list contains the respective values for the 50 participants.

Step 2: Pearson Correlation Coefficient

Pearson correlation coefficient measures the linear relationship between two continuous variables. The formula for Pearson's correlation coefficient is:

�
�
�
=
∑
(
�
�
−
�
ˉ
)
(
�
�
−
�
ˉ
)
∑
(
�
�
−
�
ˉ
)
2
∑
(
�
�
−
�
ˉ
)
2
r 
xy
​
 = 
∑(x 
i
​
 − 
x
ˉ
 ) 
2
 ∑(y 
i
​
 − 
y
ˉ
​
 ) 
2
 
​
 
∑(x 
i
​
 − 
x
ˉ
 )(y 
i
​
 − 
y
ˉ
​
 )
​
 

Where:

�
�
x 
i
​
  and 
�
�
y 
i
​
  are the individual data points from the lists hours_of_exercise and bmi.
�
ˉ
x
ˉ
  and 
�
ˉ
y
ˉ
​
  are the means of the hours_of_exercise and bmi lists respectively.
Step 3: Spearman's Rank Correlation

Spearman's rank correlation measures the monotonic relationship between two variables. It's calculated by ranking the data points and then calculating the Pearson correlation coefficient for the ranks. The formula for Spearman's rank correlation is the same as Pearson's correlation, but you use the ranks of the data points instead of the actual values.

Step 4: Comparison of Results

Once you calculate both correlation coefficients, you can compare them. Pearson's correlation coefficient is suitable for measuring linear relationships, while Spearman's rank correlation is more suitable for monotonic relationships (relationships that don't necessarily follow a linear pattern). If your data follows a linear pattern, Pearson's and Spearman's correlations should be similar. If the relationship is non-linear, Spearman's correlation might capture it better.

Remember that correlation does not imply causation. A strong correlation between exercise hours and BMI doesn't necessarily mean that exercise causes changes in BMI – other variables and factors might be at play.







Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.



To calculate the Pearson correlation coefficient between two variables, you need to use the following formula:

Pearson correlation coefficient (r) = Σ[(xᵢ - x̄)(yᵢ - ȳ)] / √[Σ(xᵢ - x̄)² * Σ(yᵢ - ȳ)²]

Where:

xᵢ represents each individual's hours spent watching television.
yᵢ represents each individual's level of physical activity.
x̄ is the mean of the hours spent watching television.
ȳ is the mean of the levels of physical activity.
Given that you have collected data on both variables from a sample of 50 participants, you will need to have the actual data points to proceed with the calculation. Once you have the data, follow these steps:

Calculate the mean (average) of hours spent watching television (x̄) and the mean of physical activity levels (ȳ).
For each participant, calculate (xᵢ - x̄) and (yᵢ - ȳ).
Square each of the differences obtained in step 2.
Calculate the sum of the squared differences for both x and y.
Calculate the sum of the product of the individual differences [(xᵢ - x̄) * (yᵢ - ȳ)].
Plug the values from steps 4 and 5 into the formula for the Pearson correlation coefficient.
Without the actual data, it's not possible for me to perform the exact calculation. If you provide the data (or a sample of it), I can help you calculate the Pearson correlation coefficient.






Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:
    
      Age(Years)                 Soft drink Preference
                                      
        25                                 Coke
        42                                 Pepsi
        37                               Mountain dew
        19                                   Coke
        31                                  Pepsi
        28                                   Coke
                        
        
        

        
Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.        
       
       
       
       To calculate the Pearson correlation coefficient between two variables, we will need the following formula:

�
=
∑
�
=
1
�
(
�
�
−
�
ˉ
)
(
�
�
−
�
ˉ
)
∑
�
=
1
�
(
�
�
−
�
ˉ
)
2
∑
�
=
1
�
(
�
�
−
�
ˉ
)
2
r= 
∑ 
i=1
n
​
 (x 
i
​
 − 
x
ˉ
 ) 
2
 ∑ 
i=1
n
​
 (y 
i
​
 − 
y
ˉ
​
 ) 
2
 
​
 
∑ 
i=1
n
​
 (x 
i
​
 − 
x
ˉ
 )(y 
i
​
 − 
y
ˉ
​
 )
​
 

Where:

�
n is the number of data points (30 in this case).
�
�
x 
i
​
  represents the number of sales calls made per day for the 
�
ith sales representative.
�
�
y 
i
​
  represents the number of sales made per week for the 
�
ith sales representative.
�
ˉ
x
ˉ
  is the mean of the 
�
x values.
�
ˉ
y
ˉ
​
  is the mean of the 
�
y values.
You'll need to calculate the means and then plug them into the formula to get the correlation coefficient 
�
r. However, I don't have access to the specific data points you collected, so I can't perform the exact calculations for you. If you provide me with the data points, I can guide you through the calculations.


```
